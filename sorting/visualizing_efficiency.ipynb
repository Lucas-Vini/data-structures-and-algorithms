# Visualizing the sorting algorithms efficiency

```python
import time
import random
import sorting
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

We want to know how the time to sort a list increases with the length of the list using different sorting algorithms. To do this, let's start by defining a function that genarates a random list with the given length and calculates the time to sort it with the chosen algorithm.

```python
def sorting_time_execution(list_length, sorting_algorithm):
    sorting_algorithms = {'bubble sort':sorting.bubble_sort,
                          'merge sort':sorting.merge_sort,
                          'quick sort':sorting.quick_sort,
                          'selection sort':sorting.selection_sort,
                          'insertion sort':sorting.insertion_sort,
                         }
    random_list = [random.randint(0,100) for i in range(list_length)]
    start_time = time.time()
    sorting_algorithms[sorting_algorithm](random_list)
    return time.time() - start_time
```

Now we define a function to return a list with the time execution to sort 10 random lists.

```python
def sorting_tests(list_length, sorting_algorithm):
    time_execution = []
    for i in range(20):
        time_execution.append(sorting_time_execution(list_length, sorting_algorithm))
    return time_execution
```

To organize it all, let's define a function that creates a data frame with the time execution lists with diferrent lists lenghts, with the chosen algorithm.

```python
def time_execution_df(sorting_algorithm):
    df = pd.DataFrame()
    for length in range(50, 800, 50):
        time_execution_tests = sorting_tests(length, sorting_algorithm)
        length = 'length=' + str(length)
        df[length] = time_execution_tests
    return df
```

Now we can genarate data about the time execution of different algorithms.

```python
sorting_algorithms = ('bubble sort', 'quick sort', 'insertion sort', 'merge sort', 'selection sort')
sorting_algorithms_dfs = {}
for sorting_algorithm in sorting_algorithms:
    sorting_algorithms_dfs[sorting_algorithm] = time_execution_df(sorting_algorithm)
```

Let's see how one of them looks like. For example, bubble sort.

```python
sorting_algorithms_dfs['bubble sort']
```

Finally, here's a comparasion between the sorting algorithms eficciency!

```python
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (8,8)
plt.title('Sorting algorithms efficiency', fontsize=20)
plt.xlabel('List length', fontsize=15)
plt.ylabel('Average execution time', fontsize=15)

for sorting_algorithm in sorting_algorithms:
    x, y = [], []
    df = sorting_algorithms_dfs[sorting_algorithm]
    for column in df.columns:
        x.append(int(column[7:]))
        y.append(df[column].mean())
    plt.plot(x, y, label=sorting_algorithm)
    plt.legend()
```

It's known that quick sort and merge sort have O(n*log(n)), and here we can visualize their high performance. And the others, all O(n), rapidly growing to big execution time.
